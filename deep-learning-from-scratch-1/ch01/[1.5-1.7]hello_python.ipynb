### 1.5 Numpy 연습

```python
import numpy as np
```

```python
x = np.array([1., 2., 3.])
print(x)
type(x)
```

```python
A = np.array([[1., 2.],\
              [3., 4.]])

print(A)
print(A.shape)
print(A.dtype)
```

Note that $$AB =\begin{bmatrix}
  8 & 8\\
  18 & 18
  \end{bmatrix} $$

```python
B = np.array([[2., 2.],\
              [3., 3.]])

A*B
```

그러니까 numpy의 *는 elementwise 곱하기임을 명심하자   
그러면 Broadcast도 elementwise로 곱해진다는 것을 알 수 있다

```python
C = np.array([10, 20])

print(A * 10)
print(A * C)
```

```python
X = np.array([[51, 52],[53,54],[55,56]])
print(X[2])
print(X[1][0])

for row in X:
  print(row)

print("------------------------------")
X_f = X.flatten()
print(X_f)

# 특정 인덱스에 있는 element들만 뽑아서 얻기
print(X[np.array([0,2])])
print(X_f[np.array([0,2,4])])
print("------------------------------")

print(X[X>53])
print(X[np.array([[True, False],
       [False,  True],
       [ True,  True]])])
```

### 1.6 Matplotlib

```python
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle = "--")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```
