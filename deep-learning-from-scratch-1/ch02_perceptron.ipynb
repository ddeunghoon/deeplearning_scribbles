### 2.2 Simple logic gates
* AND/NAND/OR gate truth table (for $x_1, x_2, y$) 쉽게 구현 가능
* 퍼셉트론 구조는 위 세 게이트 모두에서 똑같다(매개변수 값만 다르게 하여 구현 가능)

### 2.3 Implementation

```python
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  if w1*x1 + w2*x2 > theta:
    return 1
  else:
    return 0

print(AND(0,0))   # 0
print(AND(1,0))   # 0
print(AND(0,1))   # 0
print(AND(1,1))   # 1
```

```python
import numpy as np

x = np.array([0,1])
w = np.array([0.5, 0.5])

b = -0.7

np.sum(w*x) + b
```

* bias는 뉴런이 얼마나 쉽게 활성화 하느냐를 조정하는 매개변수   
* bias가 클수록 더 쉽게 활성화 할 것이다

```python
def AND(x1, x2):
  x = np.array([x1,x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  if np.sum(w*x) + b > 0:
    return 1
  else:
    return 0

print(AND(0,0))   # 0
print(AND(1,0))   # 0
print(AND(0,1))   # 0
print(AND(1,1))   # 1
```

```python
def NAND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([-0.5, -0.5])
  b = 0.6
  if np.sum(w*x) + b > 0:
    return 1
  else:
    return 0

def OR(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.2
  if np.sum(w*x) + b > 0:
    return 1
  else:
    return 0

print(OR(0,0))   # 0
print(OR(1,0))   # 1
print(OR(0,1))   # 1
print(OR(1,1))   # 1
```

### 2.5 다층 퍼셉트론
* 퍼셉트론 단층으로는 XOR 게이트를 구현할 수 없지만 다층으론 구현이 가능하다
* XOR 게이트가 NAND/OR/AND의 조합으로 구현 가능하기 때문에!

```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  return AND(s1, s2)

print(XOR(0,0))   # 0
print(XOR(1,0))   # 1
print(XOR(0,1))   # 1
print(XOR(1,1))   # 0
```

* 이론 상 2층 퍼셉트론이면 컴퓨터를 만들 수 있다
* 시그모이드 함수를 활성화 함수로 이용하면 2층 퍼셉트론으로도 임의의 함수를 표현할 수 있다
